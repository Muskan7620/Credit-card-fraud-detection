```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/creditcard.csv')
df.head()
```

```python
print(df.shape)
print(df['Class'].value_counts())  # 0 = legit, 1 = fraud
df.describe()
```

```python
# Transaction Amount Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['Amount'], bins=50, kde=True)
plt.title("Distribution of Transaction Amounts")
plt.show()

# Time-based Spending
df['Hour'] = df['Time'] // 3600
sns.boxplot(x='Hour', y='Amount', data=df)
plt.title("Hourly Spending Patterns")
plt.show()
```

```python
# Compare legit vs fraud transaction amount
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Transaction Amounts (0 = Legit, 1 = Fraud)")
plt.show()

# Fraud Percentage
fraud_rate = df['Class'].mean() * 100
print(f"Fraud Rate: {fraud_rate:.4f}%")
```

```python
features = ['Amount', 'Time']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[features])

model = IsolationForest(contamination=0.001)
df['anomaly_score'] = model.fit_predict(X_scaled)

# Mark predicted frauds
df['PredictedFraud'] = df['anomaly_score'].apply(lambda x: 1 if x == -1 else 0)

# Compare with true labels
pd.crosstab(df['Class'], df['PredictedFraud'], rownames=['Actual'], colnames=['Predicted'])
```

```python
px.scatter(df.sample(1000), x="Time", y="Amount", color="PredictedFraud", title="Predicted Fraud Patterns")
```
